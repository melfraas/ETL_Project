```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
users = [228, 127, 113, 108, 87]
x_axis = np.arange(len(users))
```

```python
# We apply align="edge" to ensure our bars line up with our tick marks
plt.bar(x_axis, users, color='r', alpha=0.80, align="center")
```

```python
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["California", "New York", "Texas", "Pennsylvania", "Illinois"])
```

```python
plt.xlim(-0.75, len(x_axis)-0.25)
```

```python
plt.ylim(0, max(users)+300)
```

```python
plt.title("Schools")
plt.xlabel("States")
plt.ylabel("Number of Schools Per State")
```
